```python
import pandas as pd
from Bio import SeqIO
import numpy as np
```

```python
ncbi_metadata = 'sequences.csv'
ncbi_fasta = 'sequences.fasta'

def combine_dates(row):
    x = row['Collection_Date']
    y = row['Release_Date']
    
    if x and pd.isna(x) == False and x not in ['','nan','NaN']:
        if len(str(x).split('-')) == 3:
            return x
        elif len(str(x).split('-')) == 2:
            return f'{x}-01'
        elif len(str(x).split('-')) == 1:
            return f'{x}-01-01'
        else:
            row['Collection_Date'] == ''
            return combine_dates(row)
    elif y and pd.isna(y) == False and y not in ['','nan','NaN']:
        if len(str(y).split('-')) == 3:
            return y
        elif len(str(y).split('-')) == 2:
            return f'{y}-01'
        elif len(str(y).split('-')) == 1:
            return f'{y}-01-01'
        else:
            return None
    else:
        return None
    
def add_dot(name):
    return name + '.1'

df = pd.read_csv(ncbi_metadata)
df['Name'] = df['Accession']

# Remove anything not TBEV
#df = df[df['Organism_Name']=='Tick-borne encephalitis virus']
df = df[df['Nuc_Completeness']=='complete']
df
```

```python
df = df[[
    'Organism_Name',
    'Name',
    'Release_Date',
    'Length',
    'Country',
    'Host',
    'Collection_Date',
    'Isolation_Source',
]]

# Add dot to name
#df['Name'] = df['Name'].apply(add_dot)

# Combine Release and Collection Date
df['New_Date'] = df.apply(combine_dates, axis=1)
df['New_Date'] = pd.to_datetime(df['New_Date'], format='%Y-%m-%d')

df
```

```python
df = df[['Organism_Name','Name','New_Date','Length','Country','Isolation_Source','Host']]
df.columns = ['Organism_Name','name','date','length','country','isolation_source','host']
df
```

```python
seqids = []
seqdesc = []
seqs = []

for seq in SeqIO.parse(ncbi_fasta, 'fasta'):
    seqids.append(seq.id)
    seqdesc.append(' '.join(seq.description.split('|')[1:]))
    seqs.append(str(seq.seq).strip())
    
seq_df = pd.DataFrame({
    'name': seqids,
    'desc': seqdesc,
    'seq': seqs
})
seq_df
```

```python
merged_df = pd.merge(df, seq_df, on='name', how='left')
merged_df
```

```python
# Remove anything not TBEV
#merged_df = merged_df[merged_df['Organism_Name']=='Tick-borne encephalitis virus']
del merged_df['Organism_Name']
merged_df
```

```python
merged_df.to_csv('CCHF_NCBI_metadata.tsv',sep='\t',index=False)
```
